### Pre-process the Data Set (normalization, grayscale, etc.)

Minimally, the image data should be normalized so that the data has mean zero and equal variance. For image data, `(pixel - 128)/ 128` is a quick way to approximately normalize the data and can be used in this project. 

Other pre-processing steps are optional. You can try different techniques to see if it improves performance. 

Use the code cell (or multiple code cells, if necessary) to implement the first step of your project.

````python
### Preprocess the data here. It is required to normalize the data. Other preprocessing steps could include 
### converting to grayscale, etc.
### Feel free to use as many code cells as needed.
#((x - np.mean(x),axis=0)/(np.std(x),axis=0)
#Normalize image data z = (x-mean(x))/std(x)

#z-score of 0 is at the mean of the distribution and a z-score of 2.0 or beyond is in the tails of the distribution.  
#A negative zâ€“score means that the original score was below the mean. 
#As the sample size becomes large, approximately half the z-scores should be negative and half of the 
#z-scores should be positive.

#https://en.wikipedia.org/wiki/Grayscale#Converting_color_to_grayscale
#Y' = 0.299 R + 0.587 G + 0.114 B 
#np.dot(rgb[...,:3], [0.299, 0.587, 0.114])

def preprocess(image_data):
    
    #Temporary array of same dimension of traffic image with 1 image channel
    #Assuming train,test and valid traffic image is of same size image_shape holds train image dimensions
    image_temp = np.ndarray((image_data.shape[0], image_shape[0], image_shape[0], 1), dtype=np.uint8)
    
    for index, colored_image in enumerate(image_data):
        gray_image = np.dot(colored_image[...,:3], [0.299, 0.587, 0.114])
      
        #Insert a new axis that will appear at the axis position 2 to avoid broadcasting error
        gray_image = np.expand_dims(gray_image, axis=2)
        image_temp[index] = gray_image
    
    #Return normalized data
    return (image_temp - np.mean(image_temp,axis=0))/np.std(image_temp, axis=0)


#Convert images to gray scale and normalize using Z-score(standardization)
X_train = preprocess(X_train)
X_valid = preprocess(X_valid)
X_test = preprocess(X_test)

#Print shape of images to confirm all dimensions are matching and channel is 1
print(X_train.shape)
print(X_valid.shape)
print(X_test.shape)

#Display image to make sure images are grayed and label are correct
index = random.randint(0, len(X_train))
v_index = random.randint(0, len(X_valid))
t_index = random.randint(0, len(X_test))

image = X_train[index].squeeze()
v_image = X_valid[v_index].squeeze()
t_image = X_test[t_index].squeeze()

plt.figure(figsize=(1,1))
#Display random training image
plt.imshow(image, cmap="gray")
print(y_train[index])


from sklearn.utils import shuffle

X_train, y_train = shuffle(X_train, y_train)
````

### Model Architecture

````python
### Define your architecture here.
import tensorflow as tf

#Assign hyperparameters some value

#Epochs indicate total number of times data is run through given network
#We need to tweak this parameters so that cost is between underfitting to overfitting
#i.e increase till validation accuracy decreases and training accuracy increases avoid overfitting
#If validation data is not provided take 20% of training data as good rule of thumb
EPOCHS = 10

#batch size indicates total training data for a given batch
BATCH_SIZE = 128

DROPOUT = 0.5 

from tensorflow.contrib.layers import flatten

#SAME padding
#out_height = ceil(float(in_height) / float(strides[1]))
#out_width = ceil(float(in_width) / float(strides[2]))

#Valid padding
#out_height = ceil(float(in_height - filter_height + 1) / float(strides[1]))
#out_width = ceil(float(in_width - filter_width + 1) / float(strides[2]))


def conv(x,sh,cs=1,k=2,s=2,p='VALID',mu = 0, sigma = 0.1):
    conv_W = tf.Variable(tf.truncated_normal(shape=sh, mean = mu, stddev = sigma))
    conv_b = tf.Variable(tf.zeros(sh[3]))
    conv_l = tf.nn.conv2d(x, conv_W, strides=[1, cs, cs, 1], padding=p) + conv_b

    conv_l =  tf.nn.relu(conv_l)

    conv_l = tf.nn.max_pool(conv_l, ksize=[1, k, k, 1], strides=[1, k, k, 1], padding=p)

    return conv_l

def full_conn(fc,sh,dropout,mu = 0, sigma = 0.1,act='X'):
    fc_W = tf.Variable(tf.truncated_normal(shape=sh, mean = mu, stddev = sigma))
    fc_b = tf.Variable(tf.zeros(sh[1]))
    fc_l   = tf.matmul(fc, fc_W) + fc_b
    
    if act == 'X':
        fc_l = tf.nn.relu(fc_l)
        return tf.nn.dropout(fc_l, dropout)
    else:
        return fc_l


def LeNet(x,dropout):    
    
    #Layer 1: Convolutional. Input = 32x32x1. Output = 28x28x6
    #Pooling. Input = 28x28x6. Output = 14x14x6.
    conv1 = conv(x, (5, 5, 1, 6))
    
    #Layer 1: Convolutional. Input = 14x14x6. Output = 10x10x16
    #Pooling. Input = 10x10x16. Output = 5x5x16.
    conv2 = conv(conv1, (5, 5, 6, 16))

    #Flatten. Input = 5x5x16. Output = 400.
    fc0   = flatten(conv2)
    
    #Layer 3: Fully Connected. Input = 400. Output = 120
    fc1 = full_conn(fc0,(400, 120),dropout)

    #Layer 4: Fully Connected. Input = 120. Output = 84
    fc2 = full_conn(fc1,(120, 84),dropout)
    
    #Layer 5: Fully Connected. Input = 84. Output = 43.
    logits = full_conn(fc2,(84, 43),dropout,act=' ')
    
    return logits
### Feel free to use as many code cells as needed.
````

### Train, Validate and Test the Model

A validation set can be used to assess how well the model is performing. A low accuracy on the training and validation
sets imply underfitting. A high accuracy on the training set but low accuracy on the validation set implies overfitting.

````python
### Train your model here.

x = tf.placeholder(tf.float32, (None, 32, 32, 1))
y = tf.placeholder(tf.int32, (None))
keep_prob = tf.placeholder(tf.float32)
one_hot_y = tf.one_hot(y, 43)

rate = 0.001

logits = LeNet(x,keep_prob)
cross_entropy = tf.nn.softmax_cross_entropy_with_logits_v2(labels=one_hot_y, logits=logits)
loss_operation = tf.reduce_mean(cross_entropy)
optimizer = tf.train.AdamOptimizer(learning_rate = rate)
training_operation = optimizer.minimize(loss_operation)

### Calculate and report the accuracy on the training and validation set.

correct_prediction = tf.equal(tf.argmax(logits, 1), tf.argmax(one_hot_y, 1))
accuracy_operation = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
saver = tf.train.Saver()

def evaluate(X_data, y_data):
    num_examples = len(X_data)
    total_accuracy = 0
    sess = tf.get_default_session()
    for offset in range(0, num_examples, BATCH_SIZE):
        batch_x, batch_y = X_data[offset:offset+BATCH_SIZE], y_data[offset:offset+BATCH_SIZE]
        accuracy = sess.run(accuracy_operation, feed_dict={x: batch_x, y: batch_y, keep_prob: 1.})
        total_accuracy += (accuracy * len(batch_x))
    return total_accuracy / num_examples
### Once a final model architecture is selected, 
### the accuracy on the test set should be calculated and reported as well.

with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    num_examples = len(X_train)
    
    print("Training...")
    print()
    for i in range(EPOCHS):
        X_train, y_train = shuffle(X_train, y_train)
        for offset in range(0, num_examples, BATCH_SIZE):
            end = offset + BATCH_SIZE
            batch_x, batch_y = X_train[offset:end], y_train[offset:end]
            sess.run(training_operation, feed_dict={x: batch_x, y: batch_y, keep_prob: 0.60})
            
        validation_accuracy = evaluate(X_valid, y_valid)
        print("EPOCH {} ...".format(i+1))
        print("Validation Accuracy = {:.3f}".format(validation_accuracy))
        print()
        
    saver.save(sess, './lenet')
    print("Model saved")
### Feel free to use as many code cells as needed.
````
